```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Pour ce faire j'ai importé une database de kaggle que voici:

```python
data = pd.read_csv("heart.csv")
data.head()
```

cp = crise cérébrale; trestbps = battements par seconde;  chol = cholesterol;    
fbs = Vitesse trop rapide du sucre dans le sang;    
restECG = surchange du coeur; thalach = rithme cardiaque maximum enregistré;
exang = perte de sang; thal = manque de fer;
target = risque potentiel de crise caridaque 


```python
homme = data[data. sex == 1]
femme = data[data. sex == 0]
target0 = data[data. target == 0]
target1 = data[data. target == 1]
```

J'ai entré celle ligne de donnée afin de séparer les hommes des femmes 

```python
plt.subplot(2,2,1)
plt.scatter(homme.cp,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(2,2,2)
plt.scatter(femme.cp,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
plt.subplot(2,2,3)
plt.scatter(homme.target,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(2,2,4)
plt.scatter(femme.target,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
```

Grace à ces tableaux, je peux en conclure que chez les hommes (les tableau de gauches) les crises cérébrales ne dépensent pas de l'âge, tandis que les femmes les moins agées ont peut de chance d'en avoir plus de trois. 
Les cibles des études de crises cardiaques couvreent toutes la palettes des âges que ce soit pour les hommes  que pour les femmes.

```python
plt.subplot(2,2,1)
plt.scatter(homme.exang,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(2,2,2)
plt.scatter(femme.exang,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
plt.subplot(2,2,3)
plt.scatter(homme.thal,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(2,2,4)
plt.scatter(femme.thal,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
```

J'ai réalisé ces deux graphiques car je voulais savoir si une insufisance de fer ou une perte de sang pouvais correspondre au graphique des crises cérébrales.
Le stade de thal (thalassémie) a l'aire de correspondre au graphique de crise cérébral car le state de thal est un stade minimum.

```python
plt.subplot(5,2,1)
plt.scatter(homme.target,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(5,2,2)
plt.scatter(femme.target,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
plt.subplot(5,2,3)
plt.scatter(homme.fbs,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(5,2,4)
plt.scatter(femme.fbs,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
plt.subplot(5,2,5)
plt.scatter(homme.restecg,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(5,2,6)
plt.scatter(femme.restecg,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
plt.subplot(5,2,7)
plt.scatter(homme.chol,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(5,2,8)
plt.scatter(femme.chol,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
plt.subplot(5,2,9)
plt.scatter(homme.trestbps,homme.age, s=10, c='blue', edgecolor='black', linewidth=1)
plt.subplot(5,2,10)
plt.scatter(femme.trestbps,femme.age, s=10, c='red', edgecolor='black', linewidth=1)
```

Ce nuage de point nous permet d'analyser la dfférence entre hommes et femmes sur les risques potentiels de crise cardiaque, les fbs, restECG, cholestérole et trest BPS.
