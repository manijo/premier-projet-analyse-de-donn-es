```python
print("Here is a chart of countries with fewer than 1,000 employees in textile finishing between 2007 and 2008.")
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2007,2008])

y1 = np.array([0, 10])
y2 = np.array([513, 506])
y3 = np.array([511, 403])
y4 = np.array([67, 61])
y5 = np.array([396, 463])
y6 = np.array([5, 5])
y7 = np.array([320, 300])
y8 = np.array([246, 243])
y9 = np.array([145, 152])
y10 = np.array([492, 550])

plt.plot(x, y1, "ko:", label="Ecuador")
plt.plot(x, y2, "co-", label="Finland")
plt.plot(x, y3, "ko-", label="Hungary")
plt.plot(x, y4, "mo-", label="Ireland")
plt.plot(x, y5, "ro-", label="Lithuania")
plt.plot(x, y6, "yo:", label="Malta")
plt.plot(x, y7, "go-", label="New Zeland")
plt.plot(x, y8, "ro:", label="Norway")
plt.plot(x, y9, "yo-", label="Slovakia")
plt.plot(x, y10, "go-", label="Sweden")

plt.legend()
plt.xlim(2007 , 2008.5)
plt.xticks( [2007 , 2008])
plt.ylim(0, 600)
plt.title("Number of employees in textile finishing")
plt.xlabel("Years")
plt.ylabel("Number of employees")

plt.show()
```
