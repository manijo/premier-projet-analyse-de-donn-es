Lien fichier data de départ : https://www.kaggle.com/lperez/coronavirus-france-dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("covid_data_france.csv")
df = df.fillna("(valeur manquante)")
```

Tri des différentes valeurs :

```python
sexe = df.sexe.value_counts()
print(sexe)
region = df.region.value_counts()
print(region)
departement = df.departement.value_counts()
print(departement)
raison_infection = df.raison_infection.value_counts()
print(raison_infection)
sante = df.sante.value_counts()
print(sante)
```

Représentation graphique patients contaminés par region :

```python
df['region'].value_counts().head(20).plot.bar()
plt.title("Nombres de patients contaminés par région")
plt.xlabel("Régions")
plt.ylabel("Nb de patients contaminés")
```

Représentation graphique patients par sexe (avec valeur manquante) :

```python
labels = 'Homme', 'Femme', 'Valeur manquante'
sizes = [102, 111, 1859]
colors = ['red', 'blue', 'green']
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.2f%%', shadow=False, startangle=90)
plt.title("Répartition des patients par sexe (avec valeurs manquantes)")
```

Représentation graphique patients par sexe (sans valeur manquante) :

```python
labels = 'Homme', 'Femme'
sizes = [102, 111]
colors = ['red', 'blue']
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.2f%%', shadow=False, startangle=90)
plt.title("Répartition des patients par sexe (sans valeurs manquantes)") 
```
